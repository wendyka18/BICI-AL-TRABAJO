```python
import numpy as np, pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from ucimlrepo import fetch_ucirepo

rng = np.random.default_rng(123)

# 1. Carga el dataset completo desde UCI
bike = fetch_ucirepo(id=275)  # Bike Sharing Dataset

# 2. Extrae las características (X) y el objetivo (y)
df = bike.data.original.copy()


# 3. Limpieza mínima y orden
df["dteday"] = pd.to_datetime(df["dteday"])
df = df.sort_values("dteday").reset_index(drop=True)

print(df.head())

# 4. Selección de columnas (incluye casual/registered)
df = df[['season','yr','mnth','holiday','weekday','workingday',
         'weathersit','temp','hum','windspeed','casual','registered','cnt','dteday']]

print(df.head(), df.columns.tolist())
```

## **PREGUNTA 1**
¿La distribución de cnt puede tratarse como aproximadamente normal, realiza el histrograma y el gráfico Q-Q?

```python
# 1. Visualización
plt.figure(figsize=(14, 5))
plt.hist(df['cnt'], bins=50)
plt.title('Distribución de bicicletas rentadas')
plt.xlabel('NUMERO DE RENTAS')
plt.ylabel('Frecuencia')
plt.show()
import statsmodels.api as sm
qqplot=stats.probplot(df['cnt'], dist="norm", plot=plt)
plt.title('Gráfico Q-Q de cnt')
plt.show()
```

## **PREGUNTA 2**
¿Cómo co-varía cnt con temp, hum y windspeed? ¿La relación es lineal o solo monótona?. Ayuda: Calcular la correlación de Pearson y de Spearman.

```python
# Construimos nuevas columnas
columnas_relacion = ['cnt', 'temp', 'hum', 'windspeed']
df_corr = df[columnas_relacion]

pearson_corr = df_corr.corr(method='pearson')
print("Pearson Correlation:")
print(pearson_corr)

spearman_corr = df_corr.corr(method='spearman')
print("\nSpearman Correlation:")
print(spearman_corr)
```

### **Interpretación pregunta 2**

Temperatura (temp) y bicicletas (cnt): Cuando hace más calor, la gente alquila más bicicletas. La relación es bastante directa y se mantiene de forma constante (es decir, es lineal y monótona).
Humedad (hum) y bicicletas (cnt): Cuando hay más humedad, la gente alquila menos bicicletas. Esta relación también es bastante directa y se mantiene de forma constante.
Velocidad del viento (windspeed) y bicicletas (cnt): La velocidad del viento no parece afectar mucho al número de bicicletas alquiladas. La relación es muy débil.

## **PREGUNTA 3**
¿Desde qué tamaño muestral el TLC justifica intervalos y pruebas basados en normalidad para la media de cnt?




### **Respuesta pregunta 3**

El Teorema del Límite Central (TLC) indica que, independientemente de la distribución de la población, la distribución de las medias muestrales se convertira a una distribución normal a medida que el tamaño de la muestra (n) aumenta.

No hay un tamaño muestral exacto universalmente  para que el TLC se aplique , ya que depende de la forma de la distribución original. para distribuciones moderadamente asimétricas , un tamaño de muestra de n ≥ 30 suele ser suficiente para que la distribución de la media muestral sea aproximadamente normal. Para distribuciones muy asimétricas, se necesitarían tamaños muestrales mayores.

## **PREGUNTA 5**
¿Cuál es un IC bootstrap para la mediana de cnt y cómo cambia tu mensaje si la distribución es sesgada?

```python
# Número de remuestreos
n_iterations = 5000
# sample_size = df['cnt'] # This was causing the error
cnt_laborales = df[df['workingday'] == 1]['cnt'] # Ensure cnt_laborales is defined
sample_size = len(cnt_laborales) # Use the length of the data as sample size
medians = []
for i in range(n_iterations):
    # REMUESTREO CON REEMPLAZO
    bootstrap_sample = cnt_laborales.sample(n=sample_size, replace=True, random_state=rng) # Use cnt_laborales as intended
    #CALCULAR LA MEDIANA DE LA MUESTRA BOOSTRAP
    median = bootstrap_sample.median()
    medians.append(median)

# CALCULAR EL INTERVALO DE CONFIANZA DEL 95%
confidence_interval = np.percentile(medians, [2.5, 97.5])

print("Intervalo de confianza bootstrap del 95% para la mediana de cnt:")
print(confidence_interval)


#Calculate the median of the original data
original_median = cnt_laborales.median() # Calculate median on cnt_laborales as intended
print("Mediana original de cnt:", (original_median))
```

Interpretación pregunta 5

el intervalo de confianza Bootstrap del 95% para la mediana de la cantidad de bicicletas rentadas está entre 148 y 155.Hay una alta probabilidad de que la mediana real de la población total de rentas de bicicletas se encuentre dentro de este rango. La mediana original de la muestra que teníamos era 151, y como este valor está justo en el centro del intervalo, nos da confianza en que nuestra estimación es precisa. Para llegar a este resultado, se tomaron 5,000 muestras virtuales de nuestros datos originales, calculando la mediana de cada una.

```python
plt.figure(figsize=(10, 6))
plt.hist(medians, bins=30, edgecolor='black')
plt.title('Distribución de las Medianas Bootstrap')
plt.xlabel('Mediana de cnt')
plt.ylabel('Frecuencia')

# Add vertical lines for the confidence interval
plt.axvline(confidence_interval[0], color='red', linestyle='dashed', linewidth=2, label=f'Límite inferior ({confidence_interval[0]:.2f})')
plt.axvline(confidence_interval[1], color='green', linestyle='dashed', linewidth=2, label=f'Límite superior ({confidence_interval[1]:.2f})')

plt.legend()
plt.show()

print("El intervalo de confianza bootstrap del 95% para la mediana de cnt es:", confidence_interval)
```
